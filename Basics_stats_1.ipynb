```python
# prompt: Calculate the mean, median, mode, and standard deviation for these
# columns in easy way

import pandas as pd
import numpy as np

sale_count=pd.read_csv('/content/sales_data_with_discounts.csv')
#print(sale_count.head())

numerical_cols = sale_count.select_dtypes(include=np.number).columns
#print(numerical_cols)

# Calculate mean, median, mode, and standard deviation for numerical columns
summary_stats = sale_count[numerical_cols].agg(['mean', 'median', 'std'])

# Calculate mode separately as agg doesn't handle it directly
for col in numerical_cols:
    mode_val = sale_count[col].mode().iloc[0]  # Get the first mode value
    summary_stats.loc['mode',col] = mode_val

summary_stats
```

```python
# prompt: Provide a brief interpretation of these statistics in less steps

import pandas as pd
import numpy as np

sale_count = pd.read_csv('/content/sales_data_with_discounts.csv')

numerical_cols = sale_count.select_dtypes(include=np.number).columns

summary_stats = sale_count[numerical_cols].agg(['mean', 'median', 'std'])

for col in numerical_cols:
    summary_stats.loc['mode', col] = sale_count[col].mode().iloc[0]

summary_stats
# Interpretation:  The table shows the mean, median, standard deviation, and mode for each numerical column in the sales dataset.
# Mean represents the average value. Median represents the middle value.
# Standard deviation measures the dispersion or spread of the data around the mean.
# Mode represents the most frequent value.  Comparing these values can give insights
# into the central tendency and variability of sales data for each feature.
```
